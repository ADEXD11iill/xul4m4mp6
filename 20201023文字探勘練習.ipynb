```python
import requests
url="https://www.chinatimes.com/money/?chdtv" ##把目標網址輸入，例子為中時財經網
res2=requests.get(url)

soup=BeautifulSoup(res2.text,'html.parser')
soup.find('h3')##印內容
res2=soup.find_all('h3')
for i in res2[0:3]:
    print(i,text)
```

```python
import requests
url="https://www.chinatimes.com/money/?chdtv" ##把目標網址輸入，例子為中時財經網
res1=requests.get(url)

soup=BeautifulSoup(res1.text,'html.parser')
soup.select('h3')##找全部
soup.select_one('h3')##h3的第一個
```
